```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import random
random.seed(777)
v = [random.randint(-5,5) for k in range(20)]
plt.figure(figsize=[12,8])
plt.title("A simple plot", size = 'x-large')
plt.ylabel('Y')
plt.xlabel('X')
plt.grid(True)
plt.plot(v, 'g-', linewidth = 6 );
```

```python
random.seed(111)
mean, std = 0, 1
X = [random.gauss(mean, std) for i in range(1500)] 
plt.xlabel('Measurement')
plt.ylabel('Probability')
plt.title('Histogram of Measurements')
plt.grid(True)
plt.hist(X, 100, density=True, facecolor='b');
```

```python
plt.subplot(222)
plt.plot(range(20), 'r+')
plt.subplot(221)
plt.plot(range(40), 'g.')
plt.subplot(224)
plt.plot(range(60), 'bx')
plt.subplot(223)
plt.plot(range(100),'c-' )
```
